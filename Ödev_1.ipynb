```python
def indoor(cevap):
  return cevap.lower()
isim = input("adınız nedir ?")
print("merhaba,",indoor(isim))
```

```python
def playback(cevap):
  return cevap.replace(" ", "...")
isim = input("adınız nedir?")
print("merhaba,"+playback(isim))
```

```python
def convert(cevap):
  cevap = cevap.replace(":)", "☺️")
  cevap = cevap.replace(":(", "🙁")
  return cevap
isim = input("Nasıl hissediyorsunuz?")
print(convert(isim))
```

```python
def einstein():
  m = int(input("Kütleyi (kg) girin:"))
  c = 300000000
  E = m * c ** 2
  print(f"Enerji (joule cinsinden): {E}")
einstein()
```

```python
def dollars_to_float(d):
    d = d.replace("$", "")
    return float(d)

def percent_to_float(y):
    y = y.replace("%", "")
    return float(y) / 100

def main():
    yemek = input("Yemek ne kadardı? ")
    yuzde = input("Yüzde kaç bahşiş bırakmak istersiniz? ")

    yemek_fiyat = dollars_to_float(yemek)
    oran = percent_to_float(yuzde)

    bahsis = yemek_fiyat * oran

    print(f"Bırakmanız gereken bahşiş: ${bahsis:.2f}")

main()
```
